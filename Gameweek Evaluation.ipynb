```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from sklearn.metrics import mean_absolute_error
from google.colab import drive
drive.mount('/content/drive')
```

```python
url = "https://fantasy.premierleague.com/api/bootstrap-static/"

response = requests.get(url)

if response.status_code == 200:
  data = response.json()
  players = data['elements']
  players_sorted = sorted(players, key=lambda x: x['id'])

else:
  print(f"Failed to retrieve data. Status code: {response.status_code}")

points_dict = {}

for player in players_sorted:
    url = 'https://fantasy.premierleague.com/api/element-summary/' + str(player['id']) + '/'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        for i in range(min(3, len(data['history']))):
            if data['history'][-(i+1)]['round'] == 36:
                current_value = points_dict.get(player['id'], 0)
                points_dict[player['id']] = current_value + data['history'][-(i+1)]['total_points']

    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
```
