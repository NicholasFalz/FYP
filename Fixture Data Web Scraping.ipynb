```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files
```

### Web Scraping Fixture Data

```python
team_stats_2017_18 = []
team_stats_2018_19 = []
team_stats_2019_20 = []
team_stats_2020_21 = []
team_stats_2021_22 = []
team_stats_2022_23 = []

%mkdir 2017-18
%mkdir 2018-19
%mkdir 2019-20
%mkdir 2020-21
%mkdir 2021-22
%mkdir 2022-23
```

#### 2017-18

```python
%cd 2017-18
```

```python
# Getting urls for all teams

url = 'https://fbref.com/en/comps/9/2017-2018/2017-2018-Premier-League-Stats'
data = requests.get(url)
soup = BeautifulSoup(data.text)

standing_table = soup.select('table.stats_table')[0]

links = standing_table.find_all('a')
links = [l.get('href') for l in links]
links = [l for l in links if '/squads' in l]
team_urls = [f'https://fbref.com{l}' for l in links]
team_urls = [url.replace('2017-2018', '2017-2018' + '/matchlogs/c9/schedule') for url in team_urls]
team_urls = [url.replace('Stats', 'Scores-and-Fixtures-Premier-League') for url in team_urls]

team_urls = sorted(team_urls, key=lambda url: url.split('/')[-1])

# Getting data from urls

for i, url in enumerate(team_urls):
  time.sleep(2)
  data = requests.get(url)
  table = pd.read_html(data.text, match = 'Scores & Fixtures')
  team_stats_2017_18.extend(table)
```

```python
for i, df in enumerate(team_stats_2017_18):
  df = df.drop(columns=['Time', 'Round', 'Day', 'Result', 'Opponent', 'Poss', 'Attendance', 'Captain', 'Referee', 'Formation', 'Match Report', 'Notes'])
  df.rename(columns = {'Date':'match_date'}, inplace = True)
  df = df.sort_values('match_date')
  df.to_csv(f'{i+1}.csv', index=False)
```

```python
%cd /content
!zip -r 2017-18.zip 2017-18/
```
